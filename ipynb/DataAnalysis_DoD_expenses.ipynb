```r
#Step 1: 
#Read data set into data.frame
AirForce <- read.csv2("AF1621.csv", header=TRUE, sep = ",")
```

```r
#Step 2:
#construct plot
AF <- ggplot((AirForce), 
aes(x = Year))
```

```r
#Step 3:
#Add details
AirForcePlot <- AF +
  geom_point(aes(x = Year,
            y = Military.Personnel/10^6, color = "Military Personnel")) +
            geom_line(aes(x = Year, y = Military.Personnel/10^6, color = "Military Personnel")) +
  geom_point(aes(x = Year,
            y = Operation...Maintenance/10^6, color = "Operation And Maintenance")) +
            geom_line(aes(x = Year, y = Operation...Maintenance/10^6, color = "Operation And Maintenance")) +
  geom_point(aes(x = Year,
            y = Procurement/10^6, color = "Procurement")) +
            geom_line(aes(x = Year, y = Procurement/10^6, color = "Procurement")) +
  geom_point(aes(x = Year, 
            y = RDT.E/10^6, color = "Research and Development")) +
            geom_line(aes(x = Year, y = RDT.E/10^6, color = "Research and Development")) +
  geom_point(aes(x = Year,
            y = Military.Constuction/10^6, color = "Construction")) +
            geom_line(aes(x = Year, y = Military.Constuction/10^6, color = "Construction")) +
  geom_point(aes(x = Year,
            y = Family.Housing/10^6, color = "Family Housing")) +
            geom_line(aes(x = Year, y = Family.Housing/10^6, color = "Family Housing")) + 
  geom_point(aes(x = Year, 
            y = Revolving...Management.Funds/10^6, color = "Revolving Management Funds")) +
            geom_line(aes(x = Year, y = Revolving...Management.Funds/10^6, color = "Revolving Management Funds")) +
  geom_point(aes(x = Year,
            y = Total/10^6)) + 
            geom_line(aes(x = Year, y = Total/10^6)) +           
  
  labs(title = "Air Force Spending over Five years (in Millions)", color = "Budget Subsection")+
  ylab("Expenditure By Title")
```

```r
OMRegression <- lm(formula = Operation...Maintenance ~ Year, data = AirForceLM)
MPRegression <- lm(formula = Military.Personnel ~ Year, data = AirForceLM)
PRRegression <- lm(formula = Procurement ~ Year, data = AirForceLM)
RDRegression <- lm(formula = RDT.E ~ Year, data = AirForceLM)
(OMRegression)
plot(AirForceLM$Operation...Maintenance,xlab = "Year", ylab = "Appropriation Title", xlim = c(0,8), ylim = c(35-000000,70000000), pch = 19)
points(y=AirForceLM$Military.Personnel, x = AirForceLM$Year, col = "blue", pch = 19)
points(y=AirForceLM$Procurement, x = AirForceLM$Year, col = "red", pch = 19)
points(y=AirForceLM$RDT.E, x = AirForceLM$Year, col = "green", pch = 19)
abline(OMRegression, lwd = 2)
abline(MPRegression, col = "blue", lwd = 2)
abline(PRRegression, col = "red", lwd = 2)
abline(RDRegression, col = "green", lwd = 2)
legend("bottomright",
       legend=c("Operation and Maintenance", "Personnel",
                "Procurement", "R&D"),
       pch=c(19, 19, 19, 19),
       lty=c(1,1,1,1),
       col=c("black", "blue", "red", "green"),
       lwd=c(2,2,2,2))
```
